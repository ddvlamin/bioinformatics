```python
from common.hmm import compute_transition, init_transition_emission, ViterbiGraph, get_seed_alignment, print_output
```

```python
def parse_input(lines):
    threshold, sigma = lines[2].strip().split(" ")
    threshold = float(threshold)
    sigma = float(sigma)
    alphabet = lines[4].strip().split(" ")
    sequences = [l.strip() for l in lines[6:]]
    return sequences, lines[0].strip(), threshold, sigma, alphabet
```

```python
input_lines = [
"DUMMY",
"------------",
"0.358 0.01",
"--------",
"A B C D E",
"--------",
"A-A",
"ADA",
"ACA",
"A-C",
"-EA",
"D-A",
]
```

```python
sequences, observations, threshold, sigma, alphabet = parse_input(input_lines)
columns2remove = get_seed_alignment(sequences, threshold)
n = len(sequences[0]) - len(columns2remove)
transition, emission = init_transition_emission(n, alphabet)
compute_transition(transition, emission, sequences, columns2remove, alphabet, sigma)
```

```python
print_output(transition, emission, alphabet)
```

```python
with open("../data/dataset_26259_5.txt", "r") as fin:
    lines = [line.strip() for line in fin]
    sequences, observations, threshold, sigma, alphabet = parse_input(["",""] + lines)
    columns2remove = get_seed_alignment(sequences, threshold)
    n = len(sequences[0]) - len(columns2remove)
    transition, emission = init_transition_emission(n, alphabet)
    compute_transition(transition, emission, sequences, columns2remove, alphabet, sigma)
    print_output(transition, emission, alphabet)
```
