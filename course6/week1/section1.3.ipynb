```python
from collections import defaultdict

def make_trie(patterns):
    trie = dict()
    node_counter = 0
    for pattern in patterns:
        node_iter = 0
        for c in pattern:
            if node_iter not in trie:
                node_counter += 1
                trie[node_iter] = {c: node_counter}
                node_iter += 1
            else:
                if c in trie[node_iter]:
                    node_iter = trie[node_iter][c]
                else:
                    node_counter += 1
                    trie[node_iter][c] = node_counter
                    node_iter = node_counter
    return trie

def iterate_trie(trie, start=0):
    for label, nid in trie.get(start,{}).items():
        print(f"{start}->{nid}:{label}")
        iterate_trie(trie, nid)
```

```python
def prefix_match(trie, text):
    node_iter = 0
    prefix = ""
    for t in text:
        if node_iter in trie:
            node_iter = trie[node_iter].get(t)
            if node_iter is None:
                return ""
            prefix += t
        else:
            return prefix
    if node_iter not in trie:
        return prefix
    else:
        return ""
```

```python
def trie_matching(text, patterns):
    trie = make_trie(patterns)
    match_positions = []
    for start in range(len(text)-1):
        m = prefix_match(trie,text[start:])
        if m:
            match_positions.append(start)
    return match_positions
```

```python
text = "AATCGGGTTCAATCGGGGT"
patterns = [
    "ATCG",
    "GGGT",
]
trie_matching(text, patterns)
```
