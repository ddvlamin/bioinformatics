```python
import math
from copy import deepcopy

sample_input = "-3 +4 +1 +5 -2"
input = [i for i in map(int, sample_input.split(" "))]
```

```python
def reverse(l, start, stop):
    if start == 0:
        reversed = l[stop::-1]
    else:
        reversed = l[stop:start-1:-1]
    l[start:stop+1] = [-i for i in reversed]

def greedy_sorting(perm):
    sorting = []
    for k in range(0, len(perm)):
        if abs(perm[k]) != k+1: 
            try:
                position = input.index(k+1)
            except ValueError:
                position = input.index(-(k+1))
            reverse(perm, k, position)
            sorting.append(deepcopy(perm))
        if perm[k] < 0:
            perm[k] = -perm[k]
            sorting.append(deepcopy(perm))
    return sorting

def print_result(reversals):
    for reversal in reversals:
        print(" ".join([f"{i:+d}" for i in reversal]))
```

```python
print_result(greedy_sorting(input))
```

```python
sample_input = "+3 +4 +5 -12 -8 -7 -6 +1 +2 +10 +9 -11 +13 +14"
input = [i for i in map(int, sample_input.split(" "))]
```

```python
def find_breakpoints(input):
    input = [0] + input + [len(input)+1]
    breakpoints = 0
    for k in range(len(input)-1):
        diff = input[k+1]-input[k]
        if diff != 1:
            breakpoints += 1
    return breakpoints
```

```python
find_breakpoints(input)
```
