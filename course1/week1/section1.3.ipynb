```python
complement_map = {
    "A": "T",
    "G": "C",
    "T": "A",
    "C": "G"
}

def reverse_complement(dna_string):
    reverse = list()
    for nucleotide in dna_string[::-1]:
        reverse.append(complement_map[nucleotide])
    return "".join(reverse)
```

```python
print(reverse_complement("ATGC"))
```

```python
with open("dataset_3_2.txt","r") as fin:
    dna_string = fin.read().strip()
```

```python
print(reverse_complement(dna_string))
```
