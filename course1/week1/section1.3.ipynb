### Question 1

```python
complement_map = {
    "A": "T",
    "G": "C",
    "T": "A",
    "C": "G"
}

def reverse_complement(dna_string):
    reverse = list()
    for nucleotide in dna_string[::-1]:
        reverse.append(complement_map[nucleotide])
    return "".join(reverse)
```

```python
print(reverse_complement("ATGC"))
```

```python
with open("dataset_3_2.txt","r") as fin:
    dna_string = fin.read().strip()
```

```python
print(reverse_complement(dna_string))
```

### Question 2

```python
def find_pattern_positions(pattern, dna_string):
    positions = []
    start_index = 0
    while start_index<=len(dna_string)-len(pattern):
        try:
            start_index = dna_string.index(pattern, start_index)
        except ValueError:
            break
        positions.append(start_index)
        start_index += 1
    return positions
```

```python
pattern = "ATAT"
dna_string = "GATATATGCATATACTT"
assert find_pattern_positions(pattern, dna_string) == [1,3,9]
```

```python
with open(f"{basepath}datasets/dataset_3_5.txt","r") as fin:
    pattern, dna_string = fin.read().strip().split("\n")
print(" ".join([str(i) for i in find_pattern_positions(pattern, dna_string)]))
```
