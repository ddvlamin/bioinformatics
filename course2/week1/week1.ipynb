```python
def composition(k, dna_string):
    kmers = []
    for i in range(0,len(dna_string)-k+1):
        kmers.append(dna_string[i:i+k])
    return set(kmers)
```

```python
assert composition(5, "CAATCCAAC") == set([
    "CAATC",
    "AATCC",
    "ATCCA",
    "TCCAA",
    "CCAAC"
])
```

```python
with open("dataset_197_3.txt","r") as fin:
    dna_string = ""
    for i, line in enumerate(fin):
        if i==0:
            k=int(line.strip())
        else:
            dna_string += line.strip()
```

```python
for kmer in composition(k, dna_string):
    print(kmer)
```
