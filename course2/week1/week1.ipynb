```python
from collections import defaultdict
```

```python
def composition(k, dna_string):
    kmers = []
    for i in range(0,len(dna_string)-k+1):
        kmers.append(dna_string[i:i+k])
    return set(kmers)
```

```python
assert composition(5, "CAATCCAAC") == set([
    "CAATC",
    "AATCC",
    "ATCCA",
    "TCCAA",
    "CCAAC"
])
```

```python
def path_to_genome(kmers):
    k = len(kmers[0])
    genome = kmers[0]
    for i, kmer in enumerate(kmers[1:]):
        prev_kmer = kmers[i]
        assert kmer[0:k-1] == prev_kmer[1:k]
        genome += kmer[-1]
    return genome
```

```python
kmers = [
    "ACCGA",
    "CCGAA",
    "CGAAG",
    "GAAGC",
    "AAGCT"
]
expected_result = "ACCGAAGCT"
```

```python
assert expected_result == path_to_genome(kmers)
```

```python
def overlap_graph(patterns):
    k = len(patterns[0])
    prefixes = defaultdict(list)
    suffixes = defaultdict(list)
    adjacency = defaultdict(list)
    
    for kmer in patterns:
        prefix = kmer[0:k-1]
        suffix = kmer[1:k]
        prefixes[prefix].append(kmer)
        suffixes[suffix].append(kmer)
        
    for suffix, kmers in suffixes.items():
        l = prefixes.get(suffix,[])
        for kmer in kmers:
            adjacency[kmer].extend(l)
        
    return adjacency

def print_overlap_graph(adjacency):
    adj_strings = set()
    for prefix, suffixes in adjacency.items():
        suffix = ",".join(suffixes)
        if suffixes:
            adj_strings.add(
                f"{prefix} -> {suffix}"
            )
    return adj_strings
```

```python
patterns = [
    "ATGCG",
    "GCATG",
    "CATGC",
    "AGGCA",
    "GGCAT",
    "GGCAC"
]
expected_results = {
    "CATGC -> ATGCG",
    "GCATG -> CATGC",
    "GGCAT -> GCATG",
    "AGGCA -> GGCAC,GGCAT"
}
```

```python
results = print_overlap_graph(overlap_graph(patterns))
print(results)
assert results == expected_results
```
