```python
from collections import defaultdict
```

```python
def composition(k, dna_string):
    return set(composition_iterator(k, dna_string))

def composition_iterator(k, dna_string):
    for i in range(0,len(dna_string)-k+1):
        yield dna_string[i:i+k]
```

```python
def path_to_genome(kmers):
    k = len(kmers[0])
    genome = kmers[0]
    for i, kmer in enumerate(kmers[1:]):
        prev_kmer = kmers[i]
        assert kmer[0:k-1] == prev_kmer[1:k]
        genome += kmer[-1]
    return genome
```

```python
kmers = [
    "ACCGA",
    "CCGAA",
    "CGAAG",
    "GAAGC",
    "AAGCT"
]
expected_result = "ACCGAAGCT"
```

```python
assert expected_result == path_to_genome(kmers)
```

```python
def overlap_graph(patterns):
    k = len(patterns[0])
    prefixes = defaultdict(list)
    suffixes = defaultdict(list)
    adjacency = defaultdict(list)
    
    for kmer in patterns:
        prefix = kmer[0:k-1]
        suffix = kmer[1:k]
        prefixes[prefix].append(kmer)
        suffixes[suffix].append(kmer)
        
    for suffix, kmers in suffixes.items():
        l = prefixes.get(suffix,[])
        for kmer in kmers:
            adjacency[kmer].extend(l)
        
    return adjacency

def print_overlap_graph(adjacency):
    adj_strings = list()
    for prefix, suffixes in adjacency:
        suffix = ",".join(sorted(suffixes))
        if suffixes:
            adj_strings.append(
                f"{prefix} -> {suffix}"
            )
    return adj_strings
```

```python
patterns = [
    "ATGCG",
    "GCATG",
    "CATGC",
    "AGGCA",
    "GGCAT",
    "GGCAC"
]
expected_results = {
    "CATGC -> ATGCG",
    "GCATG -> CATGC",
    "GGCAT -> GCATG",
    "AGGCA -> GGCAC,GGCAT"
}
```

```python
results = print_overlap_graph(overlap_graph(patterns).items())
print(results)
```

```python
def debruijn(genome, k):
    graph = defaultdict(list)
    for kmer in composition_iterator(k, genome):
        graph[kmer[0:k-1]].append(kmer[1:])
    return graph
```

```python
genome = "AAGATTCTCTAAGA"
k = 4
```

```python
result = print_overlap_graph(sorted(debruijn(genome, k).items(), key=lambda x: x[0]))
```

```python
expected = [
    "AAG -> AGA,AGA",
    "AGA -> GAT",
    "ATT -> TTC",
    "CTA -> TAA",
    "CTC -> TCT",
    "GAT -> ATT",
    "TAA -> AAG",
    "TCT -> CTA,CTC",
    "TTC -> TCT"
]
```

```python
assert expected == result
```

```python
def debruijn_from_patterns(patterns):
    k = len(patterns[0])
    graph = defaultdict(list)
    for pattern in patterns:
        prefix = pattern[0:k-1]
        suffix = pattern[1:]
        graph[prefix].append(suffix)
    return graph
```

```python
input_patterns = [
    "GAGG",
    "CAGG",
    "GGGG",
    "GGGA",
    "CAGG",
    "AGGG",
    "GGAG"
]

expected = [
    "AGG -> GGG",
    "CAG -> AGG,AGG",
    "GAG -> AGG",
    "GGA -> GAG",
    "GGG -> GGA,GGG"
]
```

```python
g = debruijn_from_patterns(input_patterns).items()
result = print_overlap_graph(sorted(g, key=lambda x: x[0]))
print(result)
```

```python
assert expected==result
```
