```python
input = """0 -> 3
1 -> 0
2 -> 1,6
3 -> 2
4 -> 2
5 -> 4
6 -> 5,8
7 -> 9
8 -> 7
9 -> 6"""
```

```python
def parse_input(input):
    adjacency_list = defaultdict(list)
    for line in input.split("\n"):
        src, dst = line.strip().split(" -> ")
        dsts = [int(i) for i in dst.split(",")]
        adjacency_list[int(src)].extend(dsts)
    return adjacency_list
```

```python
adjacency_list = parse_input(input)
print(adjacency_list)
```

```python
def print_cycle(circuit):
    #circuit += [circuit[0]]
    return "->".join([str(i) for i in circuit])
```

```python
print_cycle(eulerian_cycle(adjacency_list))
```

### k-universal string

```python
import sys
print(sys.path)
import itertools

from collections import defaultdict
from utils import *

def generate_binary_combinations(k):
    kmers = []
    for k in itertools.product('01',repeat=k):
        kmers.append("".join(k))
    return kmers
```

```python
k=3
kmers = generate_binary_combinations(k)
universal_string(kmers,'0'*(k-1))
```
